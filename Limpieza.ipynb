```python
import pandas as pd
df = pd.read_csv('AccidentesAviones.csv')
df.head()
```

# Limpieza de datos

```python
# Cambiar el nombre de unamed y convertir todas las columnas a minúsculas
df = df.rename(columns={'Unnamed: 0':'indice', 'flight_no':'n. vuelo','Ruta':'localizacion','route':'ruta','ac_type':'tipo avion', 'registration':'registro avion', 'all_aboard':'total abordo','crew_aboard':'tripulacion abordo', 'passenger_fatalities':'pasajeros fallecidos','crew_fatalities':'tripulacion fallecida','ground':'fallecidos en tierra','summary':'resumen'})
df = df.rename(columns=lambda x: x.lower())
df = df.rename(columns=lambda x: x.capitalize())

df.tail()
```

```python
# Rellenar los valores '?' con '00:00'
df['Hora declarada'] = df['Hora declarada'].replace('?', '0000')
df.head()
```

```python
df['Hora declarada'] = df['Hora declarada'].str.replace('[^\d]', '', regex=True)
df['Hora declarada'] = df['Hora declarada'].astype(str).str.zfill(4).str[:2] + ':' + df['Hora declarada'].astype(str).str.zfill(4).str[2:]

df.head()
```

```python
#Cambio de '?'por no_informa
df = df.replace('?', 'no_informa')
df_incognita = (df == '?').sum()
print(df_incognita) 
```

```python
# Formatear la fecha en el formato deseado
df['Fecha'] = pd.to_datetime(df['Fecha'])

df['Fecha'] = df['Fecha'].dt.strftime('%d-%m-%Y')

df.head()
```

```python
# Convertirmos las columnas deseadas a numerica
numerica = ['Total abordo', 'Pasajeros a bordo', 'Tripulacion abordo', 'Cantidad de fallecidos',
                      'Pasajeros fallecidos', 'Tripulacion fallecida', 'Fallecidos en tierra']

df[numerica] = df[numerica].apply(pd.to_numeric, errors='coerce')

df.dtypes
```

```python
#Ahora los valores 'no_informa' deben ser detectados como nulos en las columnas de formato float
print(df.isnull().sum())
```

```python
#Eliminamos los valores 
df = df.dropna()
```

```python
df.to_csv('datos-henry.csv', sep=';', index=False)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('Airplane_Crashes_and_Fatalities_Since_1908_20190820105639.csv')
```

```python
df = df.rename(columns={'Date':'fecha', 'Time': 'hora declarada','Location':'localizacion', 'Operator':'operador',
                         'Flight #':'n. vuelo','Route':'ruta','AC Type':'tipo avion', 'Registration':'registro avion', 
                         'Aboard':'total abordo','Aboard Passangers':'pasajeros a bordo',
                         'Aboard Crew':'tripulacion abordo', 'Fatalities Passangers':'pasajeros fallecidos',
                         'Fatalities Crew':'tripulacion fallecida','Fatalities':'cantidad de fallecidos','Ground':'fallecidos en tierra','Summary':'resumen'})
df = df.rename(columns=lambda x: x.lower())
df = df.rename(columns=lambda x: x.capitalize())

# Reemplazar los valores con formatos diferentes a 'HH:MM'
df['Hora declarada'] = df['Hora declarada'].str.replace(r'[^0-9:]', '', regex=True)
# Cambiar los valores vacios o nulos con '00:00'
df['Hora declarada'] = df['Hora declarada'].fillna('00:00')

# Valores vacios o nulos a 'no_informa'
df = df.fillna('no_informa')

# Convertir la columna de fecha a un objeto de fecha
df['Fecha'] = pd.to_datetime(df['Fecha'])

# Formatear la fecha en el formato deseado
df['Fecha'] = df['Fecha'].dt.strftime('%d-%m-%Y')
df['Fecha'] = pd.to_datetime(df['Fecha'])
df.head()
```

```python
# Convertirmos las columnas deseadas a numerica
numerica = ['Total abordo', 'Pasajeros a bordo', 'Tripulacion abordo', 'Cantidad de fallecidos',
                      'Pasajeros fallecidos', 'Tripulacion fallecida', 'Fallecidos en tierra']

df[numerica] = df[numerica].apply(pd.to_numeric, errors='coerce')

df.dtypes
```

```python
df.to_csv('datos_adicionales.csv', sep=';', index=False)
```
